```python
from elasticsearch import Elasticsearch, helpers
import json

from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()

ENDPOINT_URL = user_secrets.get_secret("ENDPOINT_URL")
USERNAME = user_secrets.get_secret("USERNAME")
PASSWORD = user_secrets.get_secret("PASSWORD")
INDEX = user_secrets.get_secret("INDEX")


# Connect to Elasticsearch Cloud
client = Elasticsearch(
    ENDPOINT_URL,
    verify_certs=False,
    basic_auth=(USERNAME, PASSWORD)
)

# Open the file and read json object
with open("/kaggle/input/yelp-dataset/yelp_academic_dataset_business.json") as f:
    data = []
    for line in f:
        try:
            doc = json.loads(line.strip())
            data.append(doc)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
            
    helpers.bulk(client, data, index=INDEX)
```
